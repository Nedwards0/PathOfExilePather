```python
from PIL import Image
import os, os.path
```

### Loading Images Data

```python
imgs = []
path = "Data/Run_2/"
valid_images = ['.png']
for f in os.listdir(path):
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid_images:
        continue
    print(f)
    imgs.append(Image.open(os.path.join(path,f)))
```
